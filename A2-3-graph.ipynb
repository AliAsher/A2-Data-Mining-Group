```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv('data/country_vaccinations.csv')
raw.info()
```

```python
df_ppl_vaccinated = raw.groupby('country').sum().sort_values('people_vaccinated', ascending=False).head(10)

plt.title(" Pleople Vaccinated - Individuals received one or more vaccines")
plt.barh(df_ppl_vaccinated.index, df_ppl_vaccinated.people_vaccinated)

plt.show
```

```python
df_total_vac_per_hundred = raw.groupby('country').max().sort_values('total_vaccinations_per_hundred', ascending=False).head(10)

plt.title("Top 10 countries that are total vaccination per hunderred")
plt.barh(df_total_vac_per_hundred.index, df_total_vac_per_hundred.total_vaccinations_per_hundred)

plt.show
```
